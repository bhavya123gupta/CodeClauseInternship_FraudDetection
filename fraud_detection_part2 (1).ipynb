```python
# Splitting the features into features and target
# Features: V1, V2, V3, V4,.........V28
# Class: 0 and 1
X = new_dataset.drop(columns='class', axis = 1)
Y = new_dataset['class']
```

```python
# Split the data into Train and Test 
X_train, X_test, Y_train, Y_test = train_test_split(X , Y , test_size = 0.2, stratify=Y, random_state=3)
# stratify=Y means classes 0 and 1 will be evenly distributes between test and train
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model training ( To check the accuracy of our model)
# By LogisticRegression
model = LogisticRegression(max_iter=10000)
```

```python
# training the logistic Regression model with training data 
model.fit(X_train, Y_train)
```

```python
# Evaluation of the model 1
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
# Importing necessary libraries for advanced classification models
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Importing models
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# Preparing the data (assuming `ccd` is the DataFrame and 'class' is the target variable)
X = ccd.drop(columns=['class'])  # Drop the target column from features
y = ccd['class']  # Target variable

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initializing the models
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss', random_state=42)
}
```

```python
# Training each model
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    print(f"{name} training completed.")
```

```python
# Evaluating each model
for name, model in models.items():
    # Predicting on the test set
    y_pred = model.predict(X_test)
    
    # Evaluation metrics
    print(f"\n{name} Results:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
