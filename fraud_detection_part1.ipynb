```python
#Importing libraries 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# loading the datset to Pandas Dataframe
ccd = pd.read_csv('creditcard.csv')
# Loading  first 5 rows of the dataset
ccd.head()
# label : class -> if 0 then a its legit/normal transaction but if its 1 then its a fraud 
```

```python
#loading last 5 transactions 
ccd.tail()
```

```python
# dataset information
ccd.info()
```

```python
# Checking for missing values in dataset
ccd.isnull().sum()
```

```python
# Distribution of legit tractions and fraudelent transactions 
ccd['class'].value_counts()
```

```python
#Seperating the data for analysis:
legit = ccd[ccd['class'] == 0]
fraud = ccd[ccd['class'] == 1]
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data 
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions 
ccd.groupby('class').mean()
```

```python
# Undersampling 
# Build a sample dataset from orgininal dataset containing similar distributution 
# of normal and fraudelent transactions
# number of legit transactions: 284315
# number of fraudelent transactions: 492
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenating 2 datasets
new_dataset = pd.concat([legit_sample, fraud], axis = 0 ) # adding row wise, axis = 1  means adding columnwise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['class'].value_counts()
```

```python
new_dataset.groupby('class').mean()
```

```python
# the above results of mean tells us the nature of dataset is not changed. the difference is still there.
# it helps and evolove our ml model to tell whether transactions is legit or fraudelent.
# this step of checking mean sample is very important.
# It helps us to check if the sample is good. difference between  mean is too much ( b/w class 0 and 1)
```

```python
# Splitting the features into features and target
# Features: V1, V2, V3, V4,.........V28
# Class: 0 and 1
X = new_dataset.drop(columns='class', axis = 1)
Y = new_dataset['class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into Train and Test 
X_train, X_test, Y_train, Y_test = train_test_split(X , Y , test_size = 0.2, stratify=Y, random_state=3)
# stratify=Y means classes 0 and 1 will be evenly distributes between test and train
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model training ( To check the accuracy of our model)
# By LogisticRegression
model = LogisticRegression(max_iter=10000)
```

```python
# training the logistic Regression model with training data 
model.fit(X_train, Y_train)
```

```python
# Evaluation of the model 1
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy of Training data: ',training_data_accuracy)
```

```python
# Accuracy of test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print(" Accuracy on test data: " ,test_data_accuracy)
```

```python
# K-Nearest neighbors 
model_1 = KNeighborsClassifier()
model_1.fit(X_train.values, Y_train.values)
```

```python
# Evaluation of model 2 
X_train_prediction_1 = model_1.predict(X_train)
training_data_accuracy_1 = accuracy_score(X_train_prediction_1, Y_train)
```

```python
print("Accuracy of Training data: ",training_data_accuracy_1)
```

```python
# Accuracy of test data
X_test_prediction_1 = model.predict(X_test)
test_data_accuracy_1 = accuracy_score(X_test_prediction_1,Y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy_1)
```

```python
# Decision Tree 
model_2 = DecisionTreeClassifier()
model_2.fit(X_train.values, Y_train.values)
```

```python
# Evaluation of model 3
X_train_prediction_2 = model_2.predict(X_train)
training_data_accuracy_2 = accuracy_score(X_train_prediction_2, Y_train)
```

```python
print("Accuracy of Training data: ",training_data_accuracy_2)
```

```python
# Accuracy of test data
X_test_prediction_2 = model.predict(X_test)
test_data_accuracy_2 = accuracy_score(X_test_prediction_2,Y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy_2)
```
