```python
# Splitting the features into features and target
# Features: V1, V2, V3, V4,.........V28
# Class: 0 and 1
X = new_dataset.drop(columns='class', axis = 1)
Y = new_dataset['class']
```

```python
# Extracting hour and day of the week from transaction time
ccd['transaction_time'] = pd.to_datetime(ccd['transaction_time'])  # Ensure the time column is in datetime format
ccd['hour_of_transaction'] = ccd['transaction_time'].dt.hour
ccd['day_of_week'] = ccd['transaction_time'].dt.dayofweek

# Displaying the hour and day of the week features
print("Hour and Day of the Week Features:")
print(ccd[['transaction_time', 'hour_of_transaction', 'day_of_week']].head())

# Calculating time since the last transaction
ccd = ccd.sort_values(['user_id', 'transaction_time'])
ccd['time_since_last_transaction'] = ccd.groupby('user_id')['transaction_time'].diff().dt.total_seconds()

#Displaying the time since last transaction feature
print("\nTime Since Last Transaction Feature:")
print(ccd[['user_id', 'transaction_time', 'time_since_last_transaction']].head(10))
```

```python
# Calculating amount to average transaction ratio per user
ccd['average_transaction_amount'] = ccd.groupby('user_id')['Amount'].transform('mean')
ccd['amount_to_avg_ratio'] = ccd['Amount'] / ccd['average_transaction_amount']

# Print the updated DataFrame
print(ccd[['user_id', 'Amount', 'average_transaction_amount', 'amount_to_avg_ratio']])
```

```python
from geopy.distance import geodesic

# Define a function to calculate distance
def calculate_distance(row):
    coords_1 = (row['latitude'], row['longitude'])
    coords_2 = (row['latitude_prev'], row['longitude_prev'])
    return geodesic(coords_1, coords_2).km

# Sort data and shift to get previous transaction coordinates
ccd['latitude_prev'] = ccd.groupby('user_id')['latitude'].shift(1)
ccd['longitude_prev'] = ccd.groupby('user_id')['longitude'].shift(1)

# Apply the distance calculation
ccd['distance_from_last'] = ccd.apply(lambda row: calculate_distance(row) if pd.notnull(row['latitude_prev']) else 0, axis=1)

#Print the updated DataFrame with relevant columns
print(ccd[['user_id', 'latitude', 'longitude', 'latitude_prev', 'longitude_prev', 'distance_from_last']])
```

```python
# Calculate user's average transaction hour and deviation from it
ccd['average_transaction_hour'] = ccd.groupby('user_id')['hour_of_transaction'].transform('mean')
ccd['hour_deviation'] = abs(ccd['hour_of_transaction'] - ccd['average_transaction_hour'])

# Print the updated DataFrame with relevant columns
print(ccd[['user_id', 'hour_of_transaction', 'average_transaction_hour', 'hour_deviation']])
Explanation:
```

```python
from sklearn.preprocessing import StandardScaler

# Selecting features to scale, including the new ones created previously
features_to_scale = ['amount_to_avg_ratio', 'time_since_last_transaction', 'distance_from_last', 'hour_deviation']

# Applying StandardScaler to the selected features
scaler = StandardScaler()
ccd[features_to_scale] = scaler.fit_transform(ccd[features_to_scale])

# Displaying the scaled features
print("Scaled Features:")
print(ccd[features_to_scale].head())
```

```python
# Adding a feature to classify if the transaction occurred on a weekend

ccd['is_weekend'] = ccd['transaction_time'].dt.dayofweek >= 5
ccd['is_weekend'] = ccd['is_weekend'].astype(int)  # Convert boolean to 1 (weekend) and 0 (weekday)

print("Weekend or Weekday Feature:")
print(ccd[['transaction_time', 'is_weekend']].head())
```

```python
# Calculate each user's median transaction amount

ccd['median_transaction_amount'] = ccd.groupby('user_id')['Amount'].transform('median')
ccd['amount_to_median_ratio'] = ccd['Amount'] / ccd['median_transaction_amount']

print("Transaction Amount Relative to Median:")
print(ccd[['user_id', 'Amount', 'median_transaction_amount', 'amount_to_median_ratio']].head())
```

```python
# Calculate rolling mean and std deviation of transaction amounts over the last 5 transactions
ccd['rolling_amount_mean'] = ccd.groupby('user_id')['Amount'].rolling(window=5).mean().reset_index(level=0, drop=True)
ccd['rolling_amount_std'] = ccd.groupby('user_id')['Amount'].rolling(window=5).std().reset_index(level=0, drop=True)

# Similarly for distance from last transaction
ccd['rolling_distance_mean'] = ccd.groupby('user_id')['distance_from_last'].rolling(window=5).mean().reset_index(level=0, drop=True)
ccd['rolling_distance_std'] = ccd.groupby('user_id')['distance_from_last'].rolling(window=5).std().reset_index(level=0, drop=True)

print("Rolling Statistics for Amount and Distance:")
print(ccd[['user_id', 'Amount', 'distance_from_last', 'rolling_amount_mean', 'rolling_amount_std', 'rolling_distance_mean', 'rolling_distance_std']].head(10))
```

```python
# Calculate the count of transactions in the last 15 minutes, 1 hour, and 24 hours for each user

ccd['transaction_count_15min'] = ccd.groupby('user_id')['transaction_time'].rolling('15min').count().reset_index(level=0, drop=True)
ccd['transaction_count_1hour'] = ccd.groupby('user_id')['transaction_time'].rolling('1H').count().reset_index(level=0, drop=True)
ccd['transaction_count_24hours'] = ccd.groupby('user_id')['transaction_time'].rolling('24H').count().reset_index(level=0, drop=True)

print("Transaction Count in Different Time Windows:")
print(ccd[['user_id', 'transaction_time', 'transaction_count_15min', 'transaction_count_1hour', 'transaction_count_24hours']].head(10))
```

```python
# Assume 'high_risk_merchants' and 'high_risk_locations' are lists of known high-risk merchants and locations

high_risk_merchants = ['merchant_1', 'merchant_2']  # Example values
high_risk_locations = ['location_1', 'location_2']  # Example values

ccd['is_high_risk_merchant'] = ccd['merchant'].isin(high_risk_merchants).astype(int)
ccd['is_high_risk_location'] = ccd['location'].isin(high_risk_locations).astype(int)

print("High-Risk Merchant or Location Tagging:")
print(ccd[['merchant', 'location', 'is_high_risk_merchant', 'is_high_risk_location']].head())
```
